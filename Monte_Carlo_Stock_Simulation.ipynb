```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 100
mu = 0.08
sigma = 0.20
T = 1
dt = 1/252
N = int(T/dt)
n_sims = 1000
```

```python
np.random.seed(42)

returns = np.random.normal(
    loc = (mu - 0.5 * sigma**2) * dt,
    scale = sigma * np.sqrt(dt) ,
    size =(N, n_sims)

)
    
```

```python
price_paths = S0 * np.exp(np.cumsum(returns, axis=0))
```

```python
plt.figure(figsize=(10,6))
plt.plot(price_paths, alpha=0.1)
plt.title("Monte Carlo Simulation of Stock Price Paths")
plt.xlabel("Trading Days")
plt.ylabel("price")
plt.show()

```
